```python
import numpy as np
import pandas as pd
df = pd.read_csv('/top2000.csv', index_col = "index")
df
# Display the DataFrame
```

```python
import re

# Assuming df is your DataFrame and 'Year of release' is the column containing the year information
def extract_number(year):
    p = re.compile(r'\d+')
    return re.search(r'\d+', year).group()

# Assuming df is your DataFrame and 'Year of release' is the column containing the year information
df['Year of release'] = df['Year of release'].apply(lambda year: extract_number(year))
```

```python
# Assuming df is your DataFrame
df.drop('Gross collection', axis=1, inplace=True)
# Display the DataFrame after dropping the column
df.drop('Description', axis=1, inplace=True)
df.drop('Star', axis=1, inplace=True)
df
```

```python
import pandas as pd

# Assuming 'df' is your DataFrame

# Convert 'Year of release' to int
df['Year of release'] = pd.to_numeric(df['Year of release'], errors='coerce', downcast='integer')

# Convert 'Votes' to int (assuming there may be commas in the numbers)
df['Votes'] = pd.to_numeric(df['Votes'].replace(',', '', regex=True), errors='coerce', downcast='integer')

# Display the data types after conversion
print(df.info())
```

```python
# Mapping dictionary
rating_mapping = {'AL': 0, 'R': 17, 'PG': 13, 'MG6': 6, 'PG-13': 13, 'G': 6, 'Unrated': None, 'M/PG': 13, 'Not Rated': None}

# Convert 'Rating' column to numerical values
df['AgeLimit'] = df['AgeLimit'].replace(rating_mapping)
df['AgeLimit'].value_counts()
```

```python
df['AgeLimit'].value_counts()
```

```python
df['Metascore'] = df['Metascore'].fillna(df['Metascore'].median())
df["AgeLimit"] = df["AgeLimit"].fillna(12)
df.info()
```

```python
df['Genre'] = df['Genre'].str.replace(r"[\[\]']", '')

genres_dummies = df['Genre'].str.get_dummies(', ')

# Concatenate the new columns with the original DataFrame
df = pd.concat([df, genres_dummies], axis=1)
df.drop('Genre', axis=1, inplace=True)
df.drop('AgeLimit', axis=1, inplace=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the color palette to blue
sns.set_palette("Greens")

cols = ['Year of release', 'Movie Rating']

# Set up a single row with three columns
n_rows = 1
n_cols = 2
fig, ax = plt.subplots(n_rows, n_cols, figsize=(n_cols * 5, n_rows * 5))

for c in range(0, n_cols):
    ax_i = ax[c]
    if cols[c] == 'Metascore Votes':
        # Use the logarithmically-scaled version for 'Metascore Votes'
        sns.histplot(data=df, x='Metascore Votes Log', kde=True, ax=ax_i)
        ax_i.set_title(f"Log-scaled {cols[c]}")
    else:
        sns.histplot(data=df, x=cols[c], kde=True, ax=ax_i)
        ax_i.set_title(f"{cols[c]}")

plt.tight_layout()
plt.show()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder


# Define transformers for numerical and categorical columns
numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])
```

```python
# Assuming you have already defined 'numerical_transformer' and 'categorical_transformer'
genre_list = [
    'Action', 'Adventure', 'Animation', 'Biography', 'Comedy', 'Crime', 'Documentary',
    'Drama', 'Family', 'Fantasy', 'Film-Noir', 'History', 'Horror', 'Music', 'Musical',
    'Mystery', 'Romance', 'Sci-Fi', 'Short', 'Sport', 'Thriller', 'War', 'Western'
]
categorical_columns = df.select_dtypes(include=['object', 'category']).columns
numerical_columns = df.select_dtypes(include=['int64', 'float64', 'int32', 'int16']).columns
numerical_columns = numerical_columns.drop(genre_list)

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_columns),
        ('cat', categorical_transformer, categorical_columns)
    ], remainder='passthrough')

# Create a pipeline with the preprocessor
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor)])

```
